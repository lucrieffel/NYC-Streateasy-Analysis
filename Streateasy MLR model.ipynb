```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt


brooklyn = pd.read_csv('brooklyn.csv')
x = brooklyn[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]
y = brooklyn[['rent']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state = 6)
#printing the dimensions of our array: the first value is the number of rows and the second is the number of columns

#fitting our multiple linear regression model
mlr = LinearRegression()

mlr.fit(x_train, y_train)
y_predicted_on_test = mlr.predict(x_test)
y_predicted_on_train = mlr.predict(x_train)


#plot our results using a scatterplot from matplotlib
plt.scatter(y_test, y_predicted_on_test)
plt.xlabel('Prices')
plt.ylabel('Predicted prices')
plt.title('Rent price vs. predicted rent price')
plt.show()

#finding our coefficients to analyze which variables carry the most weight in our regression model
print(mlr.coef_)

#evaluating our models accuracy by using residual analysis
print('Train score: ')
print(mlr.score(x_test, y_test))

print('Test score: ')
print(mlr.score(x_train, y_train))

#plotting our residual values vs our predicted values

residuals = y_predicted_on_test - y_test

plt.scatter(y_predicted_on_test, residuals, alpha=0.4)
plt.xlabel('y_predicted_on_test')
plt.ylabel('residuals')
plt.title('Residual Analysis')

plt.show()



#finding our mean absolute error to 
from sklearn.metrics import mean_absolute_error


y_true = y_train
y_pred = y_predicted_on_train

print('Mean absolute error: ')
mean_absolute_error(y_true, y_pred)




```
