```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np

brooklyn = pd.read_csv('brooklyn.csv')
manhattan = pd.read_csv('manhattan.csv')
queens = pd.read_csv('queens.csv')
```

```python
#brooklyn exploratory data analysis
brooklyn_onebr = brooklyn[brooklyn['bedrooms']<=1]
neighborhood_means_onebr = brooklyn_onebr[['neighborhood', 'rent']].groupby(['neighborhood']).mean()
mean_rent_onebr_by_neighborhood = neighborhood_means_onebr.sort_values(['rent']).reset_index()
print(mean_rent_onebr_by_neighborhood)

#creating a barplot of the average rent prices of a one bedroom apartment in brooklyn
print('Average rent of a one bedroom or studio apartment in Brooklyn')
plt.figure(figsize = (20,20))
ax = sns.barplot(x = mean_rent_onebr_by_neighborhood['neighborhood'], y = mean_rent_onebr_by_neighborhood['rent'], data = brooklyn, color = 'gray')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize = 20, horizontalalignment='right')
sns.set(style = 'whitegrid')
sns.set(font_scale=2.5)
plt.show()


brooklyn_categorical = brooklyn[['bedrooms', 'bathrooms', 'min_to_subway', 'no_fee']]

for i in brooklyn_categorical.columns: 
    categorical_num = brooklyn_categorical[i].value_counts()
    chart = sns.barplot(x = categorical_num.index, y = categorical_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation = 45)
    plt.show()

#scatter plot of correlation between square feet and rent    
plt.scatter(brooklyn[['size_sqft']], brooklyn[['rent']],alpha = 0.4)
plt.xlabel('Size(square feet)')
plt.ylabel('Rent')
plt.title('Square footage vs rent prices Brooklyn')
plt.show()

#scatter plot of correlation between distance to subway and rent
plt.scatter(brooklyn[['min_to_subway']], brooklyn[['rent']],alpha = 0.4)
plt.xlabel('Distance to subway(minutes)')
plt.ylabel('Rent')
plt.title('Distance to subway vs rent prices Brooklyn')
plt.show()

#scatter plot of correlation between building age and rent 
plt.scatter(brooklyn[['building_age_yrs']], brooklyn[['rent']],alpha = 0.4)
plt.xlabel('Building age(years)')
plt.ylabel('Rent')
plt.title('Building age vs rent in brooklyn')
plt.show() 
```
